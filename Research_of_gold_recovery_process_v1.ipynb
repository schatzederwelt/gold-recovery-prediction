```python
import pandas as pd
import numpy as np
from scipy import stats

# Графики
import matplotlib.pyplot as plt
import seaborn as sns

# Модели машинного обучения
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

# Метрики
from sklearn.metrics import mean_absolute_error, make_scorer


from sklearn.model_selection import (
    train_test_split,
    validation_curve,
    learning_curve,
    cross_val_score,
)

# Преобразования
from sklearn.preprocessing import StandardScaler, PowerTransformer
from sklearn.pipeline import make_pipeline
```

```python
# Функции, которые мы создавали в прошлых проектах


def get_target_correlation(df, target_col_name):
    """Возвращает таблицу корреляции с целевым признаком.

    1-й столбец - коэффициенты Spearman
    2-й столбец - коэффициенты Pearson
    """
    corr_table = (
        df
        # Столбец с корреляцией Спирмена
        .corr(method="spearman")[target_col_name]
        .drop([target_col_name])
        .rename("Spearman")
        .to_frame()
        # Добавляем столбец с корреляцией Пирсона(фокус на лин.зависимости)
        .join(
            df.corr()[target_col_name]
            .drop([target_col_name])
            .rename("Pearson")
            .to_frame()
        )
        .sort_values(by="Spearman")
    )
    return corr_table


def plot_learning_curve(title, model, features, target, score, train_sizes):
    """Рисует кривые валидации с помощью встроенной функции learning_curve"""

    train_sizes, train_scores, validation_scores = learning_curve(
        estimator=model,
        X=features,
        y=target,
        scoring=score,
        train_sizes=train_sizes,
        shuffle=True,
        cv=5,
    )

    # Берем метрики качества модели, полученные для  значений train_sizes
    # Вычисляем по ним среднее арифметическое для графиков кривых
    train_scores_mean = train_scores.mean(axis=1)
    validation_scores_mean = validation_scores.mean(axis=1)

    plt.style.use("seaborn")
    plt.plot(train_sizes, train_scores_mean, label="Training error")
    plt.plot(train_sizes, validation_scores_mean, label="Validation error")
    plt.ylabel(score, fontsize=14)
    plt.xlabel("Training set size", fontsize=14)
    plt.title(title, fontsize=14, y=1.03)
    plt.legend()
    plt.ylim(-12, -4)

    plt.show()
```

```python
def get_missing_values(data):
    """Возвращает таблицу с количеством и долей пропусков в датафрейме"""

    d = (
        data.isna()
        .sum()
        .rename("missing_count")
        .to_frame()
        .join(df.isna().mean().rename("missing_ratio").to_frame())
        .sort_values(by="missing_count", ascending=False)
        # Удаляем данные о столбцах, где количество пропусков = 0
        .loc[lambda x: x["missing_count"] > 0]
    )
    return d
```

Эффективность процесса получения золота из золотосодержащей руды (обогащения) оценивается

- T — доля золота в отвальных ХВОСТАХ после флотации/очистки.

Таким образом, целевые признаки для предсказания:

###  Корректность коэффициента обогащения

В п.2.2.3 мы уже обнаружили, что целевой признак `output.recovery` можно рассчитать через другие параметры в датасете.

Проверим, что эффективность обогащения рассчитана правильно:

- возьмем обучающую выборку `data_train`
- вычислим коэффициент для **чернового концентрата**
- оценим разницу между расчетами и `rougher.output.recovery` с помощью метрики `MAE`


Recovery(x 100%)  = C x (F-T) / F x (C-T)

- C — доля золота в концентрате ПОСЛЕ флотации;
- F — доля золота в сырье ДО флотации;
- T — доля золота в отвальных ХВОСТАХ после флотации.

Результат сохраним в столбце `rougher.output.recovery_calc`

```python
data_train.columns
```

```python
def calc_rougher_output_recovery(
    df, concentrate_au_col, feed_au_col, tail_au_col, output_col
):
    """Вычисляет коэффициент обогащения золота для чернового концентрата

    (conc_au * (feed_au - tail_au)) / (feed_au * (conc_au - tail_au))

    сохраняет результат в столбце output_col
    """
    df[output_col] = (
        (df[concentrate_au_col] * (df[feed_au_col] - df[tail_au_col]))
        / (df[feed_au_col] * (df[concentrate_au_col] - df[tail_au_col]))
        * 100
    )

    return df


# Создаем новый столбец "rougher.output.recovery_calc"
# Вычисляем коэффициент обогащения золота (черновой концентрат)
data_train = calc_rougher_output_recovery(
    data_train,
    "rougher.output.concentrate_au",
    "rougher.input.feed_au",
    "rougher.output.tail_au",
    "rougher.output.recovery_calc",
)


# Выводим исходный признак и рассчитанный через параметры коэффциент
data_train[["rougher.output.recovery_calc", "rougher.output.recovery"]]
```

Невооруженным глазом видно, что рассчитанные значения  практически `совпадают с исходными`.

Будем точными, **посчитаем MAE**

```python
print("Ошибка MAE расчета коэффициента обогащения rougher.output.recovery:\n")


print("{:.15f}".format(
    mean_absolute_error(
        data_train["rougher.output.recovery_calc"],
        data_train["rougher.output.recovery"])
    )
)
```

**Вывод**

Наши расчеты совпали со значениями признака `rougher.output.recovery`

Значит, в датасете **коэффициент обогащения** для процесса флотации **рассчитан верно**! И мы можем смело использовать его как целевой признак для машинного обучения.

```python
def add_date_columns(dataframe):
    '''Создает столбцы для даты, дня и времени'''

    data = dataframe.copy()

    # дата + время
    data.loc[:, "date_format"] = pd.to_datetime(data["date"])

    # день
    data.loc[:, "day"] = data.loc[:, "date_format"].dt.date

    # время
    data.loc[:, "time"] = data.loc[:, "date_format"].dt.time

    return data
```

#### Суммарная концентрация веществ 

Попробуем в этот раз исследовать суммарную концентрацию всех веществ, включая `коллоидный раствор`:

 - в сырье
 - в черновом концентрате
 - в финальном концентрате

В этот раз мы исключим результаты первого этапа очистки, оставив только финальный.

1. Поймем динамику, общая концентрация **должна расти** к финальному этапу (мы надеемся, что у нас хорошие данные)
2. Посмотрим, сохранятся ли **аномалии** в общей смеси.

```python
# Ищем и сохраняем названия параметров
# с содержанием металлов
parameters_list = []

for metal in ['ag', 'pb', 'au', 'sol']:      
    parameters_list.append('feed_' + metal)
    parameters_list.append('concentrate_' + metal)    

parameters_list
```

```python
# Посчитаем суммарную концентрацию в исходном датасете
# и построим гистограммы для каждого из 3-х этапов

legend_list = []  # cписок этапов обогащения руды для легенды

for stage_type in ['1_rougher_input', '1_rougher_output', '4_final_output']:
    
    # Отбираем набор столбцов с параметрами для суммирования
    columns = (
        df_columns
        .query('stage_type == @stage_type')
        .query('parameter_name in @parameters_list')['full_parameter_name']                    
    )    
    
    # Сохраняем сумму в выбранных столбцах
    data_full[stage_type + '_sum'] = data_full[columns].sum(axis=1)
    
    # Рисуем гистограмму
    legend_list.append(stage_type)
    data_full[stage_type + '_sum'].hist(alpha=0.5, bins=50,  figsize=(12, 6))     
      
   
plt.legend(legend_list, fontsize=12)
    
plt.title(label=f"Распределение суммарной концентрации веществ", fontsize=14)
plt.ylabel("Количество наблюдений", fontsize=12)
plt.xlabel("Суммарная концентрация", fontsize=12)  
plt.show()    
```

- Процесс `флотации` увеличивает среднюю **суммарную концентрацию** веществ, но **распределение** остается прежним.

Для поиска воспользуемся `validation_curve` и `learning_curve`, чтобы оценить сходимость моделей.

Для **кросс-валидации** результатов модели с выбранными параметрами на вооружение возьмем `cross_val_score`.

**МЕТРИКИ**

- В качестве метрики в этот раз мы возьмем кастомную метрику  - **sMAPE**

- Данную метрику используют в случаях, где как  разумно измерять ошибку `в процентах`.


Модуль разницы между ответом алгоритма и истинным значением делится на их среднее (`Симметричный средний процент отклонения`).

Запишем сразу функцию для вычисления.

```python
def smape_error_1(y_pred, y_target):
    '''Считает smape ошибку для 2-х векторов'''
    print(type(y_pred))
    error_array = (y_target - y_pred).abs() / ((y_target.abs() + y_pred.abs()) / 2)    
    smape = error_array.mean()
    return smape


def smape_error(y_pred, y_target):
    '''Считает smape ошибку для 2-х векторов Series'''
    error = np.abs(y_target - y_pred)
    mean = (np.abs(y_target) + np.abs(y_pred)) / 2
    
    smape = 1/ len(y_target) * np.sum(error / mean) * 100
    return smape

# Посчитаем smape 100% предсказаний
print("sMAPE ошибка на реальных данных:", smape_error(data_train['rougher.output.recovery'], data_train['rougher.output.recovery']))

print("sMAPE ошибка 'final.output.recovery', если подставить значения чернового концентрата:", smape_error(data_train['final.output.recovery'], data_train['rougher.output.recovery']))
```

Если мы умеем предсказывать значения для финального концентрата, то базовый результат - это подставить эти значения вместо чернового концентрата. Наша модель точно должна уметь лучше:))

- Задача в проекте **одновременно минимизировать** ошибки при оценке `rougher.output.recovery` и `final.output.recovery`.


- Поэтому итоговое качество моделей мы будем оценивать с помощью sMAPE **c весами**, где больший вес отдается  предсказаниям финального результата.

```python
def total_smape(rougher_smape, final_smape):
    '''Вычисляет итоговый sMAPE 25%-rougher, 75%-final'''
    total = rougher_smape * 0.25 + final_smape * 0.75
    return total

# make_scorer нужен чтобы использовать кастомную метрику во встроенных функциях (параметр scoring)
# greater_is_better = True если нужно максимизировать метрику (у нас задача минимизации)
smape_scorer = make_scorer(smape_error, greater_is_better=False)
```

Последним шагом подготовим датафреймы для **фиксирования результатов** (модели и их качество).
